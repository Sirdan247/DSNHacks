## Import libraries and datasets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns=None
%matplotlib inline
```

```python
train_data = pd.read_csv('Train.csv')
test_data = pd.read_csv('Test.csv')
sample_sub = pd.read_csv('SampleSubmission.csv')
```

```python
X = train_data.iloc[:,:-1]
y = train_data.default_status
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = pd.DataFrame(le.fit_transform(y), columns =['default_status'])
X["form_field47"] = le.fit_transform(X["form_field47"])
```

```python
train_data = pd.concat([X, y], axis =1)
```

## Dealing with Missing Values

```python
# Visualize missing values as a matrix 
import missingno as msno

msno.matrix(X) 
```

```python
# X['form_field47'].nunique()
```

```python
field16_20 = X.iloc[:,16:21].columns
```

```python
field34_39 = X.iloc[:,34:40].columns
```

```python
field45_47 = X.iloc[:,45:48].columns
```

```python
cat_features = []
for field in field16_20:
    cat_features.append(field)
for field in field34_39:
    cat_features.append(field)
for field in field45_47:
    cat_features.append(field)
```

```python
cat_features
```

```python
X.drop(['Applicant_ID'], axis=1, inplace=True)
```

```python
cont_features = [feature for feature in X.columns if feature not in cat_features]
```

```python
cont_features
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

ct = ColumnTransformer([
    ("imp_mode", SimpleImputer(strategy="most_frequent"), cat_features),
    ("imp_med", SimpleImputer(strategy="median"), cont_features)
])
```

```python
X_comp = pd.DataFrame(ct.fit_transform(X))
df_comp = pd.concat([X_comp, y], axis =1)
```

```python
df_comp.head()
```

```python
from sklearn.preprocessing import StandardScaler

ct = ColumnTransformer([("scaler", StandardScaler(), slice(14,50))], remainder= 'passthrough')
```

```python
X_scaled = pd.DataFrame(ct.fit_transform(X_comp))
df_scaled = pd.concat([X_scaled, y], axis =1)
```
